Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
Generate x values from -10 to 10 using np.linspace().
Compute y values for:

y= M⋅x^2
y=M⋅sin(x)
Plot both functions in a single figure:
Use different colors and line styles.
Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt

M = eval(input("Enter a value for M: "))

x = np.linspace(-10, 10, 100)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, 'r--', label="y = M * x^2")  
plt.plot(x, y2, 'b-', label="y = M * sin(x)")  

plt.title("Plot of y = M * x^2 and y = M * sin(x)")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid()
plt.show()
```

Q.2 Create a dataset of five subjects and scores.
Convert it into a Pandas DataFrame.
Plot the scores using a Seaborn bar plot with:
Different colors for each bar.
Annotations on top of each bar.
Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating dataset
data = {'Subjects': ['Math', 'Science', 'English', 'History', 'Computer'],'Scores': [85, 90, 78, 88, 95]}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Subjects', y='Scores', data=df, palette='viridis')

# Adding annotations
for p in ax.patches:
    ax.annotate(f"{p.get_height()}", (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Customizations
plt.title("Subject Scores", fontsize=14, fontweight='bold')
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, 100)

plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and then generate a dataset of 50 values using np.random.randn().
Create a 2x2 subplot layout displaying:
Line plot showing cumulative sum.
Scatter plot with random noise.
Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = 12345  
np.random.seed(roll_number)

data = np.random.randn(50)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(np.cumsum(data), 'b-', marker='o', label="Cumulative Sum")
axes[0, 0].set_title("Cumulative Sum Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].legend()
axes[0, 0].grid()

x_values = np.arange(50)
axes[0, 1].scatter(x_values, data, color='r', label="Random Noise")
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Random Value")
axes[0, 1].legend()
axes[0, 1].grid()

axes[1, 0].axis('off')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

Q.4 Download Data-set from the below link https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply ‘seaborn’ library to do the following.

Read Total profit of all months and show it using a line plot.
Read all product sales data and show it using a multiline plot.
Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)

plt.figure(figsize=(10, 5))
sns.lineplot(x=df["month_number"], y=df["total_profit"], marker="o", color="b")
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid()
plt.show()

plt.figure(figsize=(10, 5))
for column in df.columns[1:-1]:  
    sns.lineplot(x=df["month_number"], y=df[column], label=column)
plt.title("Sales Data for Different Products")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.legend()
plt.grid()
plt.show()

df.plot(kind="bar", figsize=(12, 6))
plt.title("Bar Chart of Sales Data")
plt.xlabel("Month")
plt.ylabel("Values")
plt.grid()
plt.show()
```
