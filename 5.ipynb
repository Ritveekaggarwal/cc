Q.1 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find

Sum of all elements
Sum of all elements row-wise
Sum of all elements column-wise

```python
import numpy as np

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

total_sum = gfg.sum()
row_sum = gfg.sum(axis=1)
col_sum = gfg.sum(axis=0)

print("Sum of all elements:", total_sum)
print("Sum of all elements row-wise:", row_sum)
print("Sum of all elements column-wise:", col_sum)
```

Q2.

For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find
Sorted array
Indices of sorted array
4 smallest elements
5 largest elements.
For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find
Integer elements only
Float elements only

```python
import numpy as np

# First array
array1 = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array = np.sort(array1)
sorted_indices = np.argsort(array1)
smallest_4 = np.sort(array1)[:4]
largest_5 = np.sort(array1)[-5:]

print("Sorted array:", sorted_array)
print("Indices of sorted array:", sorted_indices)
print("4 smallest elements:", smallest_4)
print("5 largest elements:", largest_5)

# Second array
array2 = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integer_elements = array2[array2 == array2.astype(int)]
float_elements = array2[array2 != array2.astype(int)]

print("Integer elements only:", integer_elements)
print("Float elements only:", float_elements)
```

Q.3 You are given a weekly sales dataset and need to perform various operations using NumPy broadcasting.

Generate your unique sales dataset:Take the sum of the ASCII values of the initials of your first and last name. Call this value X. (If your initials are A B → ASCII sum = 65+ 66 = 131 → sales = [131, 181, 231, 281, 331].) Create a NumPy array sales with values [X, X+50, X+100, X+150,X+200].
Compute your personalized tax rate as ((X % 5) + 5) / 100. Use broadcasting to apply this tax rate to each sales value.
Adjust sales based on discount: If sales < X+100, apply a 5% discount. If sales >= X+100, apply a 10% discount.
Expand sales data for multiple weeks: Create a 3×5 matrix representing three weeks of sales by stacking sales three times using broadcasting. Increase sales by 2% per week using element-wise broadcasting.

```python
import numpy as np

X = ord('A') + ord('B')  
sales = np.array([X, X+50, X+100, X+150, X+200])

tax_rate = ((X % 5) + 5) / 100
sales_after_tax = sales * (1 - tax_rate)

discounts = np.where(sales < X+100, 0.05, 0.10)  
sales_after_discount = sales_after_tax * (1 - discounts)

weekly_sales = np.vstack([sales_after_discount] * 3)  
week_factors = np.array([1.00, 1.02, 1.04]).reshape(3, 1)  
adjusted_weekly_sales = weekly_sales * week_factors  

print("Sales Data:", sales)
print("Sales After Tax:", sales_after_tax)
print("Sales After Discount:", sales_after_discount)
print("Adjusted Weekly Sales:\n", adjusted_weekly_sales)
```

Q4.Generate x values using np.linspace() from -10 to 10 with 100 points. Use each function from the list below and compute y values using NumPy:

Y = x2
Y = sin(x)
Y = ex
Y = log(|x| + 1) Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label="y = x^2")
plt.plot(x, y2, label="y = sin(x)")
plt.plot(x, y3, label="y = e^x")
plt.plot(x, y4, label="y = log(|x| + 1)")

plt.title("Function Plots")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid()
plt.show()
```
