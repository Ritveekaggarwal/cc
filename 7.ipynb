Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percenƟle of flaƩened array, for each column, for each row.
iii. Mean, Median and Standard DeviaƟon of flaƩened array, of each column, and each row

```python
array = np.array([[1, -2, 3], [-4, 5, -6]])

abs_values = np.abs(array)
percentiles = np.percentile(array, [25, 50, 75], axis=None)
percentiles_columns = np.percentile(array, [25, 50, 75], axis=0)
percentiles_rows = np.percentile(array, [25, 50, 75], axis=1)

mean_all = np.mean(array)
median_all = np.median(array)
std_dev_all = np.std(array)

mean_columns = np.mean(array, axis=0)
median_columns = np.median(array, axis=0)
std_dev_columns = np.std(array, axis=0)

mean_rows = np.mean(array, axis=1)
median_rows = np.median(array, axis=1)
std_dev_rows = np.std(array, axis=1)

print(abs_values)
print(percentiles)
print(percentiles_columns)
print(percentiles_rows)
print(mean_all, median_all, std_dev_all)
print(mean_columns, median_columns, std_dev_columns)
print(mean_rows, median_rows, std_dev_rows)
```

Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable

```python
def swap_elements(lst, pos1, pos2):
    temp = lst[pos1]
    lst[pos1] = lst[pos2]
    lst[pos2] = temp
    return lst

lst = [10, 20, 30, 40, 50]
pos1, pos2 = 1, 3
print(swap_elements(lst, pos1, pos2))
```

Q.5 Write a Python program to swap two elements in a set by converƟng it to a list first.

```python
def swap_set_elements(s, val1, val2):
    lst = list(s)
    idx1, idx2 = lst.index(val1), lst.index(val2)
    lst[idx1], lst[idx2] = lst[idx2], lst[idx1]
    return set(lst)

s = {1, 2, 3, 4, 5}
val1, val2 = 2, 4
print(swap_set_elements(s, val1, val2))
```
